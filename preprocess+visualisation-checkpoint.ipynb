```python
import seaborn as sns 
import matplotlib.pyplot as plt 

def plot_correlation_map( df ):

    corr = df.corr() # correlation among columns

    s , ax = plt.subplots( figsize =( 12 , 10 ) )   # fig, axes = plt.subplots()# figsize is a tuple of the width and height of the figure in inches
 

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True ) #palette or cmapseaborn color palette or matplotlib colormap. # colormap object that can map continuous values to colors
    s = sns.heatmap(       #heatmap display

        corr, 

        cmap = cmap,

        square=True, # the shape of each cell of the heatmap in a square

        cbar_kws={ 'shrink' : .9 }, #shrink: To change the size of the color bar


        ax=ax, 

        annot = True,   #the value will be shown on each cell of the heatmap.

        annot_kws = { 'fontsize' : 12 } # parameters of annotation; fontsize: To change the size of the font

 
        )
plot_correlation_map(df)  
    
```

```python
# the plot_correlation_map() function has as input the DataFrame, then the creation of correlation between columns and the definition of colorbar.(as desvribed above).The function uses Seaborn heatmap which aims to visualize the correlation matrix with a specific parameters(as described above).so the result is a figure of heatmap matrix of all the features (square form with the annotation above and the colorbar next to the figure). 
```
