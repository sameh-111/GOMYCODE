# most important features and their meanings:

bedrooms: Number of Bedrooms/House
bathrooms: Number of bathrooms/bedrooms
sqft_living: square footage of the home
sqft: square footage of the lot
floors: Total floors (levels) in house
waterfront: House which has a view to a waterfront
grade: overall grade given to the housing unit, based on King County grading system. 1 poor ,13 excellent
sqft_above: square footage of house apart from basement
qft_living15: Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area
sqft_lot15: lotSize area in 2015(implies-- some renovations)


# Split the dataset into a training set and a testing set. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


x = df["sqft_lot15"].values.reshape(-1,1)
y = df["price"].values

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 30)
```

#linear regression 

```python
reg = LinearRegression()
reg.fit(x_train, y_train)

predicted = reg.predict(x_test)

plt.scatter(x,y)
plt.xlabel("sqft lot15")
plt.ylabel("price")
plt.plot(x_test, predicted)
plt.show()
print("MSE", mean_squared_error(y_test, predicted))
print("R squared", r2_score(y_test, predicted))

```

```python
print("coeff a = ", reg.coef_)
```

```python
print("coeff b = ", reg.intercept_)
```

#polynomial regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = df[["grade","bathrooms"]].values
y = df.price.values
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state= 0)
```

```python
poly= PolynomialFeatures(degree = 3)
x_train_fit = poly.fit_transform(x_train)
```

```python
reg= LinearRegression()
reg.fit(x_train_fit, y_train)
```

```python
x_test_fit = poly.fit_transform(x_test)
predicted_test = reg.predict(x_test_fit)
```

```python
print("MSE : ", mean_squared_error(y_test, predicted_test))
```

```python
print("R2 : ", r2_score(y_test, predicted_test))
```
