```python
import pandas as pd
```

```python
data = pd.read_csv("CC GENERAL.csv")
data
```

```python
data.isnull().sum()
```

```python
data.drop(["BALANCE_FREQUENCY", "PAYMENTS", "MINIMUM_PAYMENTS", "PRC_FULL_PAYMENT", "TENURE"], axis = 1, inplace = True)
```

```python
data["CREDIT_LIMIT"].fillna(data["CREDIT_LIMIT"].mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data["CUST_ID"].dtype
```

```python
data["PURCHASES"].dtype
```

```python
data["CUST_ID"] = data['CUST_ID'].str.split('C', 1).str[1]
```

```python
data
```

# hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels = model.fit_predict(data)
agglomerative = pd.DataFrame(clust_labels)
agglomerative
```

# plot the clusters

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(data["BALANCE"], data["PURCHASES"], c = agglomerative, s = 50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("BALANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

# K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
Kmeans = KMeans(n_clusters = 5, random_state = 0)
Kmeans.fit(data)
labels = pd.DataFrame(Kmeans.labels_)
labels
```

```python
Kmeans.predict(data)
print(Kmeans.cluster_centers_)
```

# k best value

```python
sum_squared_distances = []
for k in range(1,10):
               km = KMeans(n_clusters = k)
               km = km.fit(data)
               sum_squared_distances.append(km.inertia_)
            
```

```python
plt.plot(k, sum_squared_distances)
plt.xlabel("k")
plt.ylabel("sum_squared_distances")
plt.title("Elbow method for optimal k")
plt.show
```
